```python
import turtle
t = turtle.Turtle()
canh = 100
goc = 90
#Hình vuông 1
t.forward(canh)
t.left(90)
t.forward(canh)
t.left(90)
t.forward(canh)
t.left(90)
t.forward(canh)
#Hình vuông 2
t.backward(canh/2)
t.left(goc+18)
t.forward(canh/2)
t.right(goc)
t.forward(canh)
t.right(goc)
t.forward(canh)
t.right(goc)
t.forward(canh)
t.right(goc)
t.forward(canh)
#Hình vuông 3
t.backward(canh/2)
t.right(goc-18)
t.forward(canh/2)
t.right(goc)
t.forward(canh)
t.right(goc)
t.forward(canh)
t.right(goc)
t.forward(canh)
t.right(goc)
t.forward(canh)
#Hình vuông 4
t.backward(canh/2)
t.right(goc-18)
t.forward(canh/2)
t.right(goc)
t.forward(canh)
t.right(goc)
t.forward(canh)
t.right(goc)
t.forward(canh)
t.right(goc)
t.forward(canh)
#Hình vuông 5
t.backward(canh/2)
t.right(goc-18)
t.forward(canh/2)
t.right(goc)
t.forward(canh)
t.right(goc)
t.forward(canh)
t.right(goc)
t.forward(canh)
t.right(goc)
t.forward(canh)
turtle.done()
```
