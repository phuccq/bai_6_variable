```python
import turtle
t = turtle.Turtle()
canh = 200
canh_tang = 40
t.fillcolor('green')
t.begin_fill()
t.penup()
#Vẽ hình vuông trong cùng
t.backward(canh/2)
t.pendown()
t.pencolor('red')
t.left(90)
t.forward(canh/2)
t.right(90)
t.forward(canh)
t.right(90)
t.forward(canh)
t.right(90)
t.forward(canh)
t.right(90)
t.forward(canh/2)
#Vẽ hình vuông ở giữa
t.left(90)
t.penup()
t.forward(canh_tang)
t.pendown()
t.pencolor('yellow')
t.right(90)
t.forward(canh/2 + canh_tang)
t.right(90)
t.forward(canh + canh_tang*2)
t.right(90)
t.forward(canh + canh_tang*2)
t.right(90)
t.forward(canh + canh_tang*2)
t.right(90)
t.forward(canh/2 + canh_tang)

#Vẽ hình vuông ngoài cùng
t.left(90)
t.penup()
t.forward(canh_tang)
t.pendown()
t.pencolor('blue')
t.right(90)
t.forward(canh/2 + canh_tang*2)
t.right(90)
t.forward(canh + canh_tang*4)
t.right(90)
t.forward(canh + canh_tang*4)
t.right(90)
t.forward(canh + canh_tang*4)
t.right(90)
t.forward(canh/2 + canh_tang*2)
t.end_fill()

turtle.done()
```
